```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
shp_data=pd.read_csv('shopping_data.csv')
shp_data.head()
```

```python
features=shp_data.iloc[:,3:].values
features[0:5]
```

```python
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10,7))
plt.title('customer dendogram')
dend= shc.dendrogram(shc.linkage(features,method='ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster=AgglomerativeClustering(n_clusters=5,linkage='ward',affinity='euclidean')
cluster.fit_predict(features)
```

```python
plt.figure(figsize=(10,7))
plt.scatter(features[:,0],features[:,1],c=cluster.labels_,cmap='rainbow')
```
